```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Create dfs

total_por_ano = pd.read_csv("analise_bolsas//total_por_ano.csv")
total_por_orientador = pd.read_csv("analise_bolsas//total_por_orientador.csv")
total_por_unidade = pd.read_csv("analise_bolsas//total_por_unidade.csv")
total_por_vertente = pd.read_csv("analise_bolsas//total_por_vertente.csv")
```

```python
# plot bolsas por ano
total_por_ano.plot(x="Edital", y="Total de Bolsa", kind="bar", title="Total de Bolsas por Ano")
```

```python
# plot bolsas por ano por unidade para o edital 2022
total_por_unidade_2022 = total_por_unidade[total_por_unidade["Edital"] == 2022]
total_por_unidade_2022_sum = total_por_unidade_2022.groupby("Unidade")["Total de Bolsa"].sum()
# sorted plot by y-values
total_por_unidade_2022_sum.sort_values().plot(figsize = (10, 10), kind="bar", title="Total de Bolsas por Unidade - Edital 2022")
```

```python
# plot bolsas por ano por unidade para o edital 2024
total_por_unidade_2024 = total_por_unidade[total_por_unidade["Edital"] == 2024]
total_por_unidade_2024_sum = total_por_unidade_2024.groupby("Unidade")["Total de Bolsa"].sum()
# sorted plot by y-values
total_por_unidade_2024_sum.sort_values().plot(figsize = (10, 10), kind="bar", title="Total de Bolsas por Unidade - Edital 2024")
```
